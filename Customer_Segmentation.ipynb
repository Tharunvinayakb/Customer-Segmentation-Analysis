```python
import pandas as pd

data = {
    'CustomerID': range(1, 21),
    'Age': [22, 25, 47, 52, 46, 56, 23, 27, 32, 40, 29, 35, 48, 50, 26, 31, 38, 45, 28, 33],
    'AnnualIncome': [25000, 30000, 90000, 100000, 85000, 120000, 27000, 32000, 45000, 70000,
                     35000, 50000, 95000, 110000, 28000, 42000, 60000, 80000, 34000, 48000],
    'SpendingScore': [80, 75, 20, 15, 25, 10, 85, 78, 60, 40,
                      72, 55, 18, 12, 82, 58, 45, 30, 77, 52]
}

df = pd.DataFrame(data)

df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Features for clustering
X = df[['AnnualIncome', 'SpendingScore']]

# Create 3 customer segments
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)

# Display data with clusters
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    df['AnnualIncome'],
    df['SpendingScore'],
    c=df['Cluster']
)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df['AnnualIncome'],
    df['SpendingScore'],
    c=df['Cluster']
)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")

plt.savefig("customer_segmentation.png")
plt.show()
```
